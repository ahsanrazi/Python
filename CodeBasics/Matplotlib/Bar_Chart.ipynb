# Bar Chart

```python
# In Bar Chart each of the value is represented as a vertical bar.

# In finance world the Bar Chart are useful to represent the measures like Revenue
```

```python
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
```

```python
# For Example: we heve the data of revenue of US tech companies as a lists

company = ['Google', 'Amazon', 'Microsoft', 'Meta']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
# For creating the Bar Chart of the above lists (Company and Revenue) we use bar()
# bar() doesn't contain the String values

# For this we have to create x-position array

xpos = np.arange(len(company))
print(xpos)
```

```python
plt.bar(xpos,revenue)
```

```python
# In the above graph in the x-axis we have value
# To change the values with the labels we have to use the xticks() function. It take two arguments ypos list 
# and the company list. Tt will change the company name with the corresponding value 

plt.xticks(xpos,company)

plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)

plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue)
```

```python
# We can also plot two different measures like Revenue and Profit

plt.xticks(xpos,company)

plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue, label='Revenue')
plt.bar(xpos,profit, label = 'Profit')

plt.legend()
```

```python
# In the above chart the profit and revenue are plotting on the top of each other but
# In finance the the people are plotting the bar-charts side by side so that they can get a comparison

# The reason for plotting the chart on the top of each other because the x-position is same of
# both of the plots

# We can make it different by subtracting some number from it

plt.xticks(xpos,company)

plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue, label='Revenue')
plt.bar(xpos+0.2,profit, label = 'Profit')

plt.legend()

# The reason of making the x-position array as the np.array because the adding/subtracting the number is easier
# If we create the python list and we add/subtract the number it will create the new array and add/subtract from it  
```

```python
# We can also comtrol the width of the bars by using width = -- function

plt.xticks(xpos,company)

plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue, width= 0.4, label='Revenue')
plt.bar(xpos+0.2,profit, width= 0.4, label = 'Profit')

plt.legend()
```

```python
# We can also make Horizontal Bar Charts
# For this we have to use barh()

plt.yticks(xpos,company)

plt.xlabel('Revenue')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2,revenue, label='Revenue')
plt.barh(xpos+0.2,profit, label = 'Profit')

plt.legend()
```
