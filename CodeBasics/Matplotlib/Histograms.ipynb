# Histograms

```python
# Histograms are the Bar Charts where the x-axis contains the frequency of a variable
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# For Example: We have data of blood-sugar level of different patients in a hospital and we want to know:
# How many patients are normal?
# How many of them are pre-diabetic?
# How many of them are diabetic? 

# In this case this type of Bar Chart could be useful where the bars represent different blood-sugar ranges 
# and y-axis represent the number of patients who are within that range 
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
# Working with histograms we need only single dimension array and the y-axis will contain the frequency of those numbers. 

# For plotting the Histogram

plt.hist(blood_sugar)
```

```python
# These are the Y-axis numbers
# (array([3., 3., 1., 0., 1., 1., 0., 2., 0., 2.])

# These are the bins
#  array([ 77. ,  84.3,  91.6,  98.9, 106.2, 113.5, 120.8, 128.1, 135.4,
#         142.7, 150. ]) 
#  <BarContainer object of 10 artists>)
```

```python
# By default it will plot 10 bins 
# Bins or Bucket is range 

# If we want three bins we have to give the bin parameter

plt.hist(blood_sugar, bins=3)
```

```python
# We can also change the width of the bars for this we have to provide rwidth = --
# rwidth = -- is a relative width of a bar compare to a bin size (The maximum value is 1)

plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
# By default it select the bin sizes on its own by doing some equal distribution
# we can also customize the range and change the bin parameter to specify the range explicitly

# 80 to 100 is normal, 100 to 125 is pre-diabatic and anything more than 125 is diabetic
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95) 
```

```python
# We can also change the color of the chart

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='green') 
```

```python
# We can also change the histogram type

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='green',histtype='step') 
```

```python
# For Example: we have data of sugar-levels of both men and women

# We can plot two types of histograms 
# For this we have to provide list

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange']) 
```

```python
# We can also give Histogras labels and set the Legend

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])

plt.legend() 
```

```python
# We can also set the axis labels and title of the chart

plt.xlabel('Sugar Range')
plt.ylabel('Total number of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['Men','Women'])

plt.legend()
```

```python
# We can also plot the Histograms Horizontally
# For this we use orientation = 'horizontal'

plt.xlabel('Sugar Range')
plt.ylabel('Total number of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['Men','Women'],orientation='horizontal')

plt.legend()
```
