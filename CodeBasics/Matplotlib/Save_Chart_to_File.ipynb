# Save Chart to a file

```python
# We can save chart to PNG file using the savefig() function
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet Bill','Car','Other utilities']
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0.2,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0.2,0])
plt.savefig('piechart.png', bbox_inches='tight',transparent = True)
plt.show()

# bbox_inches='tight' -> save the chart properly without chagnging the shape (height and width)
# transparent = True -> save the chart as transparent
# pad_inches = 2 -> add extra padding around the chart for proper look
```

```python
# We can also save chart as pdf file

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0.2,0])
plt.savefig('piechart.pdf', bbox_inches='tight',transparent = True)
plt.show()
```
