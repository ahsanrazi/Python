```python
# Matplotlib is a 2D plotting library that is primarily use for data visulaization.

# The visulaization helps in easily understanding the data, because the humman brain are wired to process
# images better than the text
```

```python
import matplotlib.pyplot as plt        # pyplot is the main plot object that we get in matplotlib module
%matplotlib inline   

# %matplotlib inline it can render the chart fine                  
```

```python
# For example we have two python lists X and Y
# X has number of week days and y has the temperature information associated with it

X = [1,2,3,4,5,6,7]
Y = [50,52,51,67,45,34,63]
```

```python
# For plotting the above lists we have to use plot()
# plot(x-axis, y-axis)

plt.plot(X,Y)

# x-axis has week days and y-axis has temperature
```

```python
# we can change the colour of the line

plt.plot(X,Y, color = 'green')
```

```python
# We can change the line-width

plt.plot(X,Y, color = 'green', linewidth = 5)
```

```python
# We can change the design of the line

plt.plot(X,Y, color = 'green', linewidth = 5, linestyle = 'dotted')
```

```python
# We can change the design of the line

plt.plot(X,Y, color = 'green', linewidth = 5, linestyle = 'dashdot')
```

```python
# We can also give the chart x and y label and title

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('weather')

plt.plot(X,Y, color = 'green', linewidth = 5, linestyle = 'dashdot')
```
