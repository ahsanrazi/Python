# Format Strings in plot function

```python
# Format String is used to control visual aspects of the 2D plot such as color, marker style and line style etc. 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# For example we have two python lists X and Y
# X has number of week days and y has the temperature information associated with it

X = [1,2,3,4,5,6,7]
Y = [50,52,51,67,45,34,63]
```

```python
# Simple way of plotting the 2D chart is 

plt.plot(X,Y)
```

```python
# We can control the visual properties of the above chart by specifying format string as a third argument

# Change the color to be green and marker to be plus (+)

plt.plot(X,Y, 'g+')  # g maeans green, + means plus(+) marker
```

```python
# The of the plot to dashed line we use

plt.plot(X,Y, 'g+--')
```

```python
# We can also do dash-dot

plt.plot(X,Y, 'g+-.')
```

```python
# The order of the format string doesn't matter it will work

plt.plot(X,Y, '+-.r')
```

```python
# For Diamond marker

plt.plot(X,Y, 'rD')
```

```python
# If we don't want to give format string we can also explicitly specify the keyword argument

plt.plot(X,Y, color = 'red' , marker = 'D', linestyle = '')
```

```python
# For the color we can also specify the hexadecimal combination

plt.plot(X,Y, color = '#008840' , marker = 'D', linestyle = '')
```

```python
# We can also control the marker size

plt.plot(X,Y, color = '#008840' , marker = 'D', linestyle = '', markersize=10)
```

```python
# We can also control the transparency using alpha.
# The scale of alpha is between 0 and 1 

# 0 means not visible 
# 1 means completely visible

plt.plot(X,Y, color = '#008840' , marker = 'D', linestyle = '', alpha = 0.5)
```
