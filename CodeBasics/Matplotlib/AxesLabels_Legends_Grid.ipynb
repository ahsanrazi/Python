# Axes Labels, Legend and Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# For example we have a lists that represents the weather data 

days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,49,46,45]
min_t = [43,42,40,33,35,37,38]
avg_t = [45,48,48,46,40,42,41]  
```

```python
# For plotting the chart we have to give the x-axis and y-axis

# The chart between Days and Maximum temperature

plt.plot(days,max_t)
```

```python
# In the same chat we can plot three diffeent lines

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# In the above chart the x-axis and the y-axis are missing and anyone looking at the graph will not have the 
# idea the x-axis representing the days and y-axis representing the temperature

# To set the axis labels and title of the chart

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# The chart looks better but we don't know what these lines represent, for this we have to set the Legend

# For setting the Legend we have to set the label of each of these lines

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend()
```

```python
# We can also change the location of the legend for this we have to use loc='--' in legend


plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc='lower left')
```

```python
# By default the the loc='' property is set as 'best'. Matplotlib decide where to put the legend

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc='best')
```

```python
# We can also set the shadow property of the legend

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc='best', shadow = True)
```

```python
# We can also change the font-size of the legend

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc='best', shadow = True, fontsize = 'large')
```

```python
# we can also change the Canvas of the graph to Grid (for visualizing the data points better)

# By default it is blank

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc='best', shadow = True,)

plt.grid()
```
