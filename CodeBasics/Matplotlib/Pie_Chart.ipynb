# Pie Chart

```python
# For Example we have a data of home expenses but when its come to knowing how much percentage or a pie individual
# item is taking out of my total expense plotting a Pie Chart gives the bigger help
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet Bill','Car','Other utilities']
```

```python
# For plotting the Pie Chart we use pie(values,labels)

plt.pie(exp_vals,labels=exp_labels)
```

```python
# For making the pie chart from all the sides So that it looks like perfect circle 

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
```

```python
# For reducing all the text above the pie chart

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
# We can also increase the size of pie chart (by default it is 1)

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
plt.show()
```

```python
# we can also show the percentage of each item

# we use autopct='%--f%%' -> -- we provide number of decimal points of the percentage 
# 0.0 mean no decimal points

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%')
plt.show()
```

```python
# We can also set the shadow of the chart

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.0f%%',shadow=True)
plt.show()
```

```python
# We can also take the pies out of the chart

# For this we will use explode = [] -> in the list we give the pies that we want to take out of the chart 
# We have five pies in the chart we will put zeros in the order
# [0,0,0,0,0] -> this means no pie is out of the chart

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',explode=[0,0,0,0,0])
plt.show()
```

```python
# For taking the food pie out of the chart it is in the second place we change the zero in the second place

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0,0])
plt.show()
```

```python
# We can take out many pies 

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0.2,0])
plt.show()
```

```python
# By default the pie goes in the anti-clockwise direction and its angle is 0-degree

# We can also change the angle of the pie chart where it will start
# For this we use startangle = --

# Change the angle at 45-degree

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0,0],startangle=45)
plt.show()
```

```python
# Change the angle at 180-degree

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0,0],startangle=180)
plt.show()
```
